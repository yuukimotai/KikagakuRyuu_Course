```python
df_head = df - df.mean()# 中心化
df.head(5)
```

```python
df_head.describe()
```

```python
import matplotlib.pyplot as plt
x = df_head["x"]
y = df_head["y"]
plt.scatter(x, y)# 実測値
plt.show()
```

### 傾きaの計算式
$$a = \dfrac{\displaystyle{\sum_{n=1}^{N}x_{n}y_{n}}}{\displaystyle{\sum_{n=1}^{N}x_{n}^2}}$$

```python
xx = x * x
xy = x * y
a  = xy.sum() / xx.sum()
a
```

```python
plt.scatter(x, y, label="y")
plt.plot(x, a * x, color="red" , label="y_hat")
plt.legend()
plt.show()
```

$$y - {\bar y} = a(x - {\bar x})$$
$$ → y = a(x - {\bar x}) + {\bar y} $$

```python
# 予測値の計算
x_new = 40
```

```python
mean = df.mean()
```

```python
xc = x_new - mean["x"]
yc = a * xc + mean["y"]
yc
```

```python
#　予測値を計算する関数の作成
def predict(x):
    xc = x - mean["x"]
    yc = a * xc + mean["y"]
    return yc
```

```python
predict(30)
```
